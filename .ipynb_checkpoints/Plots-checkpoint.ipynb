```python
from scipy.integrate import ode
y0, t0 = 100, 1
r = ode(dy_dx).set_integrator('zvode', method='bdf')
r.set_initial_value(y0,t0)
t1 = 500
dt = 1
ys = []
xs = []
while r.successful() and r.t < t1:
    ys.append(r.integrate(r.t + dt)[0])
    xs.append(r.t+dt)

    
plt.xlim(10,500)
plt.ylim(20,100)
plt.loglog(xs,ys)
    
```
